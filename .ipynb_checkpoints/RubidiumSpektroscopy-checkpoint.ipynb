```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from IPython.display import clear_output
from random import randint
```

```python
df = pd.read_csv(r'XFEL_KW0_Results_2.csv', header=None)

# find constant columns
const_arr = []
for i in range(len(df.columns)):
    if df.std()[i] == 0:
        const_arr.append(i)
# remove constant columns
df.drop(df.columns[const_arr], axis=1, inplace=True)

# normalization
#df = (df-df.mean())/df.std()
df = (df-df.min())/(df.max()-df.min())
train, test = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
```

```python
num_inputs = 7
num_outputs = 5
x_train, y_train = train.iloc[:,:num_inputs], train.iloc[:,num_inputs:]
x_test, y_test = test.iloc[:,:num_inputs], test.iloc[:,num_inputs:]
```

```python
# Model creation
def create_model(hidden_depth = 1, neurons=[16], actv_fct='relu'):
    model = keras.Sequential([
        keras.Input(shape=(num_inputs))] + [
        keras.layers.Dense(neurons[i], activation=actv_fct) for i in range(hidden_depth)
        ] + [
        keras.layers.Dense(num_outputs)]
    )
    
    print(model.summary())

    model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0005),
                  loss=keras.losses.MeanSquaredError(),
                  metrics=['accuracy']
    )
    return model
```

```python
checkpoint_filepath = r'.\tmp\checkpoint\weights.hdf5'
# Evaluate the saved model
best_model = create_model(6, [256,128,64,32,16,16])
best_model.load_weights(checkpoint_filepath)
#best_model = tf.keras.models.load_model(checkpoint_filepath)
best_model.evaluate(x_test, y_test)

```
